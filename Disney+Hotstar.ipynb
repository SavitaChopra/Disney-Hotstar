```python
orders["released_at"] = pd.to_datetime(orders["released_at"])
```

```python
orders["Month"] = orders["released_at"].dt.strftime("%B")
```

```python
orders
```

```python
plt.figure(figsize= (6,3))
sns.countplot(x = "Month", data = orders, color = "b")
plt.xticks(rotation = 45)
plt.show()
```

```python
orders["Hour"] = orders["released_at"].dt.strftime("%H")
orders["Hour"] = orders["Hour"]
orders = orders.sort_values(by = "Hour" )
sns.lineplot(x = "Hour", y = "imdb_rating", data = orders, estimator = "sum")
plt.show()
```
